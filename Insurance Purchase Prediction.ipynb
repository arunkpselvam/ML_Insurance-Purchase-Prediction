```python
import pandas as pd
import numpy as np
import warnings
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report
warnings.filterwarnings("ignore")
```

```python
insdf = pd.read_csv("CarInsurance.csv")
```

```python
insdf.head()
```

```python
insdf.info()
```

```python
#Checking for missing values in the dataset
insdf[insdf.isna().any(1)]
```

```python
insdf.groupby(["Purchase"]).count()
```

```python
insdf = insdf.drop(labels = "ID", axis = 1)
```

```python
array = insdf.values
X = array[:,0:85]
Y = array[:,85]
test_size = 0.30 # taking 70:30 training and test set
seed = 10  # Random numbmer seeding for reapeatability of the code
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=test_size, random_state=seed)
type(X_train)
```

```python
#Method to generate Confusion matric and Classfication Report
def evaluate_model(y_test,y_predict):
    cm = confusion_matrix(y_test, y_predict)
    cr = classification_report(y_test, y_predict)
    return cm,cr
```

```python
# LogisticRegression
Lmodel = LogisticRegression()
Lmodel.fit(X_train, y_train)
y_predict = Lmodel.predict(X_test)
Lmodel_score = Lmodel.score(X_test, y_test)
Lcm,Lcr = evaluate_model(y_test,y_predict)
print("Accuracy Score : {}".format(Lmodel_score))
print("Confusion Matrix : \n{}".format(Lcm))
print("Classification Report : \n{}".format(Lcr))
```

```python
# Naive Bayes
Nmodel=GaussianNB()
Nmodel.fit(X_train,y_train)
y_predict=Nmodel.predict(X_test)
Nmodel_score = accuracy_score(y_test,y_predict)
Ncm,Ncr = evaluate_model(y_test,y_predict)
print("Accuracy Score : {}".format(Nmodel_score))
print("Confusion Matrix : \n{}".format(Ncm))
print("Classification Report : \n{}".format(Ncr))
```

```python
# KNeighborsClassifier
NNH = KNeighborsClassifier(n_neighbors= 21 , weights = 'uniform', metric='euclidean')
NNH.fit(X_train, y_train)
y_predict = NNH.predict(X_test)
Kmodel_score = accuracy_score(y_test,y_predict)
Kcm,Kcr = evaluate_model(y_test,y_predict)
print("Accuracy Score : {}".format(Kmodel_score))
print("Confusion Matrix : \n{}".format(Kcm))
print("Classification Report : \n{}".format(Kcr))
```

```python
# Support Vector Machine
svmModel = svm.SVC(kernel = 'linear', C = 100)
svmModel.fit(X_train,y_train)
y_predict = svmModel.predict(X_test)
Smodel_score = accuracy_score(y_test,y_predict)
Scm,Scr = evaluate_model(y_test,y_predict)
print("Accuracy Score : {}".format(Smodel_score))
print("Confusion Matrix : \n{}".format(Scm))
print("Classification Report : \n{}".format(Scr))
```
